```python
customers_data = [
(1, 'Vinay', 'Bangalore'),
(2, 'Rahul', 'Mumbai'),
(3, 'Ankit', 'Delhi'),
(4, 'Sneha', 'Pune'),
(5, 'Kiran', 'Chennai'),
(6, 'Amit', 'Hyderabad')
]


df_customer = spark.createDataFrame(customers_data,["customer_id", "customer_name", "city"])
```

```python
orders_data = [
(101, 1, 500),
(102, 1, 300),
(103, 2, 700),
(104, 7, 400),
(105, None, 200),
(106, 3, 150),
(107, 8, 900)
]

df_orders = spark.createDataFrame(orders_data, ["order_id", "customer_id", "order_amount"])
```

```python
payments_data = [
(1001, 101, 'SUCCESS'),
(1002, 102, 'FAILED'),
(1003, 103, 'SUCCESS'),
(1004, 108, 'SUCCESS'),
(1005, 106, 'FAILED')
]
df_payments_data = spark.createDataFrame(payments_data,["payment_id", "order_id", "status"])
```

```python
df_inner = df_customer.join(df_orders, df_customer.customer_id == df_orders.customer_id ,"inner")
display(df_inner)
```

```python
df_left = df_customer.join(df_orders, df_customer.customer_id == df_orders.customer_id ,"left")
display(df_left)
#There is no difference between LEFT JOIN and LEFT OUTER JOIN. The keyword OUTER is optional; both return all records from the left table and matching records from the right table.
```

```python
df_left_semi = df_customer.join(df_orders, df_customer.customer_id == df_orders.customer_id ,"left_semi")
display(df_left_semi)
#filters rows from the left table based on existence in the right table.
```

```python
df_left_anti = df_customer.join(df_orders, df_customer.customer_id == df_orders.customer_id ,"left_anti")
display(df_left_anti)
```

```python
df_left_outer = df_customer.join(df_orders, df_customer.customer_id == df_orders.customer_id ,"left_outer")
display(df_left_outer)
```

```python
df_right = df_customer.join(df_orders, df_customer.customer_id == df_orders.customer_id ,"right")
display(df_right)
```

```python
from pyspark.sql.functions import broadcast
df_broadcast = df_customer.join(broadcast(df_orders), df_customer.customer_id == df_orders.customer_id ,"inner")
display(df_broadcast)
```

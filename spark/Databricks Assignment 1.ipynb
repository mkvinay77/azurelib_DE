## Question 5
Create a new column called 'RevenueCategory' that categorizes total revenue as 'Low' if less than $100,000, 'Medium' if between $100,000 and $500,000, and 'High' if greater than $500,000.

```python
# Your code here
from pyspark.sql.functions import when

df5 = df.withColumn(
    "RevenueCategory",
    when(col("TotalRevenue") < 100000, "Low")
    .when((col("TotalRevenue") >= 100000) & (col("TotalRevenue") <= 500000), "Medium")
    .otherwise("High")
)
display(df5)
```

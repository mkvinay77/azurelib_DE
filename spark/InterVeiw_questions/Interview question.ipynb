```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

# Sample data
data = [
    (101, "Vinay", 50000),
    (102, "Rahul", 70000),
    (103, "Anu", 60000),
    (104, "Kiran", 70000)
]

columns = ["emp_id", "name", "salary"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Window specification
window_spec = Window.orderBy(df["salary"].desc())

# Apply dense rank
df_rank = df.withColumn("rank", dense_rank().over(window_spec))

# Get second highest salary
second_highest = df_rank.filter(df_rank["rank"] == 2)

display(second_highest)
```

```python
second_highest.select("salary").distinct().show()
```

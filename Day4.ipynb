```python
#Imagine that you are working as developer at cricinfo website.
#There is a match going and you have to take the runs scored by
# each player in the team.
# At the end display the entire score card where you display run of each player
# along with team total runs.


#Assume we have 5 player in a team
# When you display the score card, also tell me how many players get out on 0.

players = []
sum = 0
i = 0
totalZeroScored = 0
while i < 5:
  p1= int(input("Enter the runs scored ="))
  players.append(p1)
  sum = sum + p1
  i = i+1

  if p1 == 0:
    totalZeroScored = totalZeroScored+1


for i in players:
  print(i)
print(sum)
print("total players who scored zero",totalZeroScored)


###logic
# In one line (simple English)
# totalZeroScored = totalZeroScored+1
# “Every time a player scores 0 runs, increase the zero-score count by 1.”
```

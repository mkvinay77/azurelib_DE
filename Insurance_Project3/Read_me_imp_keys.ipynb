✅ What does True do?

The last parameter is:

recurse = True
👉 Meaning:
True → move all files + subfolders recursively
False (default) → only moves if it's a simple file (not a directory)
#🔥 In simple words
Value	Behavior
True	Move entire folder + everything inside it
False	Fails if it's a folder (or only moves single file)
📦 Your case

You're moving:

landing/PolicyData/  →  processed/PolicyData/<timestamp>/

Since PolicyData/ is a folder, you MUST use:

True

Otherwise ❌ it will fail.

⚠️ Important Note
mv() = move (cut + paste), not copy

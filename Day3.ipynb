```python
#Imagine that you are bulding the Zepto app.
#User can add the products to the cart.
#your task is to ask the product price which is added to the cart and find the total order value.
#Assume max cart item size is 3
#While calculating the total, remeber Zepto is offering the new year discount
#when the total oder value is above 500 you get shipping charges is 0 rs otherwise shipping charges is 100 rupes


i = 0
sum =0
while i < 3:
  a1 = (int(input("Enter the iteam price = ")))
  sum = sum + a1
  i = i + 1
print("The total amount is",sum)

discount = 0
if sum > 500:
  discount = sum - 100
else:
  discount = sum + 100

total_sum = discount

print(total_sum)
```
